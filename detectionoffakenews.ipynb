```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake=pd.read_csv("C:\\Users\\Angsh\\Downloads\\Fakenews.csv")
data_true=pd.read_csv("C:\\Users\\Angsh\\Downloads\\Truenews.csv")
```

```python
data_fake["class"]=0
data_true["class"]=1
```

```python
data_fake.shape,data_true.shape
```

```python
data_merge =pd.concat([data_fake,data_true],axis=0)
data_merge.head()
```

```python
data_merge.columns
```

```python
data= data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()  # convert to lowercase
    text = re.sub(r'\[.*?\]', '', text)  # remove text in square brackets
    text = re.sub(r'\\W', ' ', text)  # replace non-word characters with space
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # remove URLs
    text = re.sub(r'<.*?>+', '', text)  # remove HTML tags
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # remove punctuation
    text = re.sub(r'\n', '', text)  # remove newline characters
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()  # typo fixed here
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt=LR.predict(xv_test)
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB=GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
pred_gb=GB.predict(xv_test)
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rf))
```
